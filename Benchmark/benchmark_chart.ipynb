```python
# %matplotlib inline
# import matplotlib
# matplotlib.use('TKAgg')
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define functions to display values on a column
def autolabel_for_accuracy_chart(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width(), 1.03*height, '%s' % float(height))
```

```python
fig_1, ax_1 = plt.subplots()

width = 0.25
index = np.arange(2)

data_llvm = [0.60358, 0.61203]
data_mxnet = [0.60358, 0.61203]
data_cuda = [0, 0.61203]

ax1 = ax_1.bar(index, data_mxnet, width=width, label='mxnet')
ax2 = ax_1.bar(index + width, data_llvm, width=width, label='tvm target=llvm')
ax3 = ax_1.bar(index + 2 * width, data_cuda, width=width, label='tvm target=cuda')

autolabel_for_accuracy_chart(ax1)

ax_1.set_ylabel('Accuracy(100%)')
ax_1.set_ylim((0, 1))
ax_1.set_title("Accuracy of inference in subset 'tiger cat' 1424p")
ax_1.set_xticks(index + width)
ax_1.set_xticklabels(('MacBook(x86)','JetsonNano(ARM)'))
ax_1.legend()
fig_1.tight_layout()
# plt.show()
plt.savefig('Accuracy_of_inference_cat.png', dpi=300)
```

```python
# Define functions to display values on a column
def autolabel_for_time_table(rects):
    for rect in rects:
        height = rect.get_height()
        if height != 0:
            plt.text(rect.get_x() + 0.02, 1.01*height, '%s' % float(height))
```

```python
fig_3, ax_3 = plt.subplots()

width = 0.25
width_bar = 0.25
index = np.arange(2)

time_llvm = [66.331, 135.575]
time_mxnet = [75.289, 804.978]
time_cuda = [0, 86.743]

ax1 = ax_3.bar(index, time_mxnet, width=width_bar, label='mxnet')
ax2 = ax_3.bar(index + width, time_llvm, width=width_bar, label='tvm target=llvm')
ax3 = ax_3.bar(index + 2 * width, time_cuda, width=width_bar, label='tvm target=cuda')

autolabel_for_time_table(ax1)
autolabel_for_time_table(ax2)
autolabel_for_time_table(ax3)

ax_3.set_ylabel('Totally cost of time(sec)')
ax_3.set_ylim((0, 900))
ax_3.set_title("Inference time in subset 'tiger_cat' 1424p")
ax_3.set_xticks(index + width)
ax_3.set_xticklabels(('MacBook(x86)','JetsonNano(ARM)'))
ax_3.legend()
fig_3.tight_layout()
# plt.show()
plt.savefig('Inference_time_cat.png', dpi=300)
```
